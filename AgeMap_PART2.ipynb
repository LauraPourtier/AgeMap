```python
# Select the columns for ages over 65 and over 80
age_columns = ['_65_69_anys', '_70_74_anys', '_75_79_anys', '_80_84_anys', '_85_89_anys', '_90_94_anys', '_95_anys_i_m__s']
super_age_columns = ['_80_84_anys', '_85_89_anys', '_90_94_anys', '_95_anys_i_m__s']

# Create a new column by summing values in selected age columns
soledad['elderly_alone'] = soledad[age_columns].sum(axis=1)
soledad['super_elderly_alone'] = soledad[super_age_columns].sum(axis=1)

# Change datatype to INT64
soledad['elderly_alone'] = pd.to_numeric(soledad['elderly_alone'])

# Create a mapping dictionary from 'cens' dataframe
district_mapping = cens.set_index('codi_districte')['nom_districte'].to_dict()

# Add 'nom_districte' column to 'soledad' using the mapping
soledad['nom_districte'] = soledad['codi_districte'].map(district_mapping)

# Exclude the row where the neighborhood is 'barcelona'
soledad = soledad[soledad['nom_barri'] != 'barcelona']

# Filter rows where the 'codi_districte' is not '<NA>'
soledad = soledad[soledad['codi_districte'].notna()]

# Display the result
soledad
```

## Total persones / Total Health Centers // BARRI

```python
# Calculate total population per neighborhood
population_per_neighborhood = cens.groupby('nom_barri')['Valor'].sum().reset_index(name='total_population')

# Count the number of health centers per neighborhood
health_centers_per_neighborhood = health_centers.groupby('addresses_neighborhood_name').size().reset_index(name='health_center_count')

# Merge the population and health center counts based on neighborhood
merged_df = pd.merge(population_per_neighborhood, health_centers_per_neighborhood, how='left', left_on='nom_barri', right_on='addresses_neighborhood_name')

# Calculate the ratio of people per health center per neighborhood
merged_df['ratio_people_per_health_center'] = merged_df['total_population'] / merged_df['health_center_count']

# Display the result
merged_df[['nom_barri', 'total_population', 'health_center_count', 'ratio_people_per_health_center']]
```

```python
# Plotting
plt.figure(figsize=(18, 6))
plt.bar(merged_df['nom_barri'], merged_df['ratio_people_per_health_center'], color=colors)
plt.xlabel('Neighborhood')
plt.ylabel('Ratio of People per Health Center')
plt.title('Ratio of People per Health Center per Neighborhood in Barcelona')
plt.xticks(rotation=90)  # Adjust rotation for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

## Total persones / Total Health Centers // DISTRICTE

```python
# Calculate total population per district
population_per_district = cens.groupby('nom_districte')['Valor'].sum().reset_index(name='total_population')

# Count the number of health centers per district
health_centers_per_district = health_centers.groupby('addresses_district_name').size().reset_index(name='health_center_count')

# Merge the population and health center counts based on district
merged_district_df = pd.merge(population_per_district, health_centers_per_district, how='left', left_on='nom_districte', right_on='addresses_district_name')

# Calculate the ratio of people per health center per district
merged_district_df['ratio_people_per_health_center'] = merged_district_df['total_population'] / merged_district_df['health_center_count']

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(merged_district_df['nom_districte'], merged_district_df['ratio_people_per_health_center'], color=colors)
plt.xlabel('District')
plt.ylabel('Ratio of People per Health Center')
plt.title('Ratio of People per Health Center per District in Barcelona')
plt.xticks(rotation=45, ha='right')  # Adjust rotation for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

## Total persones / Total Health Centers // BARRI // TIPUS CENTRE
