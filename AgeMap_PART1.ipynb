#TOTAL HEALTH CENTERS

```python
# Get the total number of health centers
total_health_centers = df_health_centers.shape[0]

print(f'The total number of health centers is: {total_health_centers}')
```

```python
#Number of health centers by types
df_health_centers_types = df_health_centers.groupby('type')['primary_key'].count().reset_index()
df_health_centers_types = df_health_centers_types.rename(columns={'primary_key': 'count of health centers'})
df_health_centers_types
```

```python
df_health_centers.dtypes
```

```python
#CREATE A DATAFRAME COUNT HEALTH CENTERS BY DISTRICT

# Create a pivot table to get counts of each type in each district
df_pivot_district = df_health_centers.pivot_table(index='addresses_district_name', columns='type', values='primary_key', aggfunc='count', fill_value=0, margins=True, margins_name='Total')

# Reset the index to make 'district' a regular column
df_pivot_district.reset_index(inplace=True)

# Rename the 'Total' column to 'Total_health_centers'
df_pivot_district.rename(columns={'Total': 'Total_health_centers','addresses_district_name': 'nom_districte'}, inplace=True)
df_pivot_district = df_pivot_district.sort_values('Total_health_centers',ascending=False)
df_pivot_district
```

```python
#CREATE A DATAFRAME COUNT HEALTH CENTERS BY BARRI

# Create a pivot table to get counts of each type in each barri
df_pivot_barri = df_health_centers.pivot_table(index='addresses_neighborhood_name', columns='type', values='primary_key', aggfunc='count', fill_value=0, margins=True, margins_name='Total')

# Reset the index to make 'district' a regular column
df_pivot_barri.reset_index(inplace=True)

# Rename the 'Total' column to 'Total_health_centers'
df_pivot_barri.rename(columns={'Total': 'Total_health_centers','addresses_neighborhood_name': 'nom_barri'}, inplace=True)

df_pivot_barri
```

#POPULATION LEAVING ALONE

```python
#DATA PREPARATION

# Select the columns for ages over 65 and over 80
age_columns = ['_65_69_anys', '_70_74_anys', '_75_79_anys', '_80_84_anys', '_85_89_anys', '_90_94_anys', '_95_anys_i_m__s']
super_age_columns = ['_80_84_anys', '_85_89_anys', '_90_94_anys', '_95_anys_i_m__s']

# Exclude the row where the neighborhood is 'barcelona'
df_soledad = df_soledad[df_soledad['nom_barri'] != 'barcelona']

# Create a mapping dictionary from 'cens' dataframe
district_mapping = df_cens.set_index('codi_districte')['nom_districte'].to_dict()

# Add 'nom_districte' column to 'soledad' using the mapping
df_soledad['nom_districte'] = df_soledad['codi_districte'].map(district_mapping)

# Create a new column by summing values in selected age columns
df_soledad['elderly_alone'] = df_soledad[age_columns].sum(axis=1)
df_soledad['super_elderly_alone'] = df_soledad[super_age_columns].sum(axis=1)

df_soledad.head()
```

```python
# Normalize the values to get percentages
df_percentage = df_merged[['Rest of the population', 'Population_Above_65', 'Population_Above_80']].div(df_merged['Total_Population'], axis=0) * 100

plt.figure(figsize=(14, 6))
bar_width = 0.5  # Width of each bar
bar_positions = range(len(df_merged['nom_districte']))

# Plotting stacked bars
plt.bar(bar_positions, df_percentage['Rest of the population'], width=bar_width, label='Rest of the population', color='skyblue')
plt.bar(bar_positions, df_percentage['Population_Above_65'], width=bar_width, bottom=df_percentage['Rest of the population'], label='Population Above 65', color='orange')
plt.bar(bar_positions, df_percentage['Population_Above_80'], width=bar_width, bottom=df_percentage['Rest of the population'] + df_percentage['Population_Above_65'], label='Population Above 80', color='green')

# Adding labels and title
plt.xlabel('District')
plt.ylabel('Population Percentage')
plt.title('Percentage Population Distribution by Age Group and District')
plt.xticks(bar_positions, df_merged['nom_districte'])
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
